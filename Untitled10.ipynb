```python
# Diviser le dataset en train/test :

X = df1['commentaire']
Y = df1['emotion']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y)
```

```python
# Vectorisation TF - IDF :

tfidf = TfidfVectorizer(max_features = 5000, stop_words="english")
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
# Entrainement d'un modele (ex : Logistic Regression) :

clf = LogisticRegression(max_iter = 1000)
clf.fit(X_train_tfidf, Y_train)
```

```python
# Evaluation :

Y_pred = clf.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
# Copy du dataset :

df2 = df1.copy()
```

```python
# Télechargement des stopwords :

nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
```

```python
# Création du fonction de nétoyage du text :

def clean_text(text):
    stemmer = PorterStemmer()
    text = re.sub("[^a-zA-Z]", " ", text)
    text = text.lower()
    text = text.split()
    text = [stemmer.stem(word) for word in text if word not in stopwords]
    return " ".join(text)
```

```python
# Application de clean_text au dataset :

df2['cleaned_commentaire'] = df2['commentaire'].apply(clean_text)
```

```python
# Division du dataset df2  :

X_train, X_test, Y_train, Y_test = train_test_split(df2['cleaned_commentaire'],
                                                    df2['emotion'],
                                                    test_size = 0.2, 
                                                    random_state = 42
                                                   )
```

```python
# Afficher les valeurs séparer :

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("Y_train:", Y_train.shape)
print("Y_test:", Y_test.shape)
```

```python
# Appliquée tfidfvectorizer :

tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Définition des classes :

classifiers = {
    "Multinomial Naive Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
}
```

```python
# Création d'une boucle d'entrainement et d'evaluation :

for name, clf in classifiers.items():
    print(f"\n===== {name}=====")
    clf.fit(X_train_tfidf, Y_train)
    Y_pred_tfidf = clf.predict(X_test_tfidf)
    accuracy_tfidf = accuracy_score(Y_test, Y_pred_tfidf)
    print(f"\naccuracy de TF-IDF: {accuracy_tfidf}")
    print("classifiication report:")
    print(classification_report(Y_test, Y_pred_tfidf))
```

```python
# Je retiens la Logistic Regression :

lg = LogisticRegression()
lg.fit(X_train_tfidf, Y_train)
```

```python
# Prediction et evaluation:

lg_Y_pred = lg.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(Y_test, lg_Y_pred))
print(classification_report(Y_test, lg_Y_pred))
```

```python
# Sauvegarder le modéle et le vecteur TF-IDF pour utilisation future :

# Sauvegarder le modele
with open("logistic_model.pkl", "wb") as f:
     pickle.dump(lg, f)

# Sauvegarder le TF_IDF
with open("tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(tfidf_vectorizer, f)

# Sauvegarder le Label-Encoder

with open("Label_Encoder.pkl", "wb") as f:
    pickle.dump(lb, f)
```

```python
# Création du fonction de prédiction d'emotion:

def predict_emotion(input_text):
    cleaned_text = clean_text(input_text)
    input_vectorized = tfidf_vectorizer.transform([cleaned_text])

    predicted_label = lg.predict(input_vectorized)[0]
    predicted_emotion = lb.inverse_transform([predicted_label])[0]
    label = np.max(lg.predict(input_vectorized))

    return predicted_emotion, label
```

```python
# Exemple de prediction :

sentences = [
    "i didnt feel humiliated",
    "i feel strong and good overall",
    "im grabbing a minute to post i feel greedy wrong",
    "He was speechles when he found out he was accepted to this new job",
    "This is outrageous, how can you talk like that?",
    "I feel like im all alone in this world",
    "He is really sweet and caring",
    "You made me very crazy",
    "i am ever feeling nostalgic about the fireplace i will know that it is still on the proporty",
    "i am feeling grouchy",
    "He hates you"
]

for sentence in sentences:

    print(sentence)
    pred_emotion, label = predict_emotion(sentence)
    print("prediction :",pred_emotion)
    print("label :",label)
    print("================================================================================")
```
